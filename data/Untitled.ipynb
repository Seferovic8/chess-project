```python
import numpy as np
import glob
import pandas as pd
import time
from numpy import loadtxt
from io import StringIO  
```

```python
def concat(df, sec_df):
    df=pd.concat([df,sec_df])
    df.index=np.arange(0,len(df))
    return df
```

```python
df=None
for i,name in enumerate(glob.glob('../26*.csv')):
    if i ==0:
        df = pd.read_csv(name,index_col=0)
    else:
        df=concat(df,pd.read_csv(name,index_col=0))
```

```python
def string_to_array(string):
    string = string.replace("]", "").replace("[", "").replace("\n","").replace(" ","").replace("."," ")
    #k = '['+k+']'
    c = StringIO(string)

    ar=loadtxt(c)
    return ar.reshape((8,8,12))
```
